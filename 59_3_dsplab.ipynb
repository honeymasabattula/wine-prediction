atleast 10 graphs on the dataset -- different and relevant types required --


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#----------------------- VISUALIZATION FOR RED WINE---------------------#
```

```python
#counting the quality ratings
sns.countplot(x='quality', data=file1)
plt.title('Count Plot of Wine Quality Ratings')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
# Example for alcohol content
sns.histplot(file1['alcohol'], bins=30, kde=True)
plt.title('Distribution of Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.show()
```

```python
#heat map of winecharacteristics
plt.figure(figsize=(12, 8))
sns.heatmap(file1.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#scatter plot of alcohol content vs quality
sns.scatterplot(x='alcohol', y='quality', data=file1)
plt.title('Scatter Plot of Alcohol Content vs Quality')
plt.xlabel('Alcohol Content')
plt.ylabel('Quality')
plt.show()
```

```python
# Example for alcohol content across different quality ratings
sns.violinplot(x='quality', y='alcohol', data=file1)
plt.title('Violin Plot of Alcohol Content by Quality Rating')
plt.xlabel('Quality')
plt.ylabel('Alcohol Content')
plt.show()
```

```python
#bar plot between fied acidity by quality rating
sns.barplot(x='quality', y='fixed acidity', data=file1)
plt.title('Bar Plot of Fixed Acidity by Quality Rating')
plt.xlabel('Quality')
plt.ylabel('Fixed Acidity')
plt.show()
```

```python
# box plots of every characteristic by qulity
for column in file1.columns[:-1]:  
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='quality', y=column, data=file1)
    plt.title(f'Box Plot of {column} by Quality Rating')
    plt.show()
```

```python
#line graphs of average wine characteristics by quality
quality_means = file1.groupby('quality').mean().reset_index()
plt.figure(figsize=(14, 8))
for column in file1.columns[:-1]:  # Exclude 'quality' column
    sns.lineplot(x='quality', y=column, data=quality_means, label=column)

plt.title('Average Wine Characteristics by Quality')
plt.xlabel('Quality')
plt.ylabel('Average Value')
plt.legend()
plt.show()
```

```python
#bar graph of average wine characteristics by quality
quality_means_melted = quality_means.melt(id_vars='quality', var_name='Characteristic', value_name='Average')
plt.figure(figsize=(14, 8))
sns.barplot(x='quality', y='Average', hue='Characteristic', data=quality_means_melted, palette='viridis')
plt.title('Average Wine Characteristics by Quality')
plt.xlabel('Quality')
plt.ylabel('Average Value')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.show()
```

```python
#pair plot of selected wine characteristics such as alcohol,ph and residual sugar by quality 
subset = file1[['alcohol', 'pH', 'residual sugar', 'quality']]
# Create a pair plot
sns.pairplot(subset, hue='quality', palette='viridis', markers=["o", "s", "D", "^", "v", "p"])
plt.suptitle('Pair Plot of Selected Wine Characteristics by Quality', y=1.02)
plt.show()
```

#  Model
Use any number ML models(regression or classification based on the problem)  of your interest, use relevant metrics  also -- - find better features of interest in given data--  20 marks


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import classification_report, accuracy_score
```

```python
#feature selection
file1['good_quality']=[1 if x>=5 else 0 for x in file1['quality']]
file2['good_quality']=[1 if x>=5 else 0 for x in file2['quality']]
```

```python
X = file1.drop(['quality','good_quality'], axis=1)
y = file1['good_quality']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#logistic regression for red wine
from sklearn.linear_model import LogisticRegression
# Train the model
logreg = LogisticRegression(max_iter=10000)
logreg.fit(X_train, y_train)
logreg.score(X_test,y_test)
```

```python
#logistic regression for white wine
X2 = file2.drop(['quality','good_quality'], axis=1)
y2 = file2['good_quality']

# Split the data
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.3, random_state=42)
from sklearn.linear_model import LogisticRegression
# Train the model
logreg = LogisticRegression(max_iter=100000)
logreg.fit(X2_train, y2_train)
logreg.score(X2_test,y2_test)
```

```python
#decision tree for red wine
from sklearn.tree import DecisionTreeClassifier

# Train the model
tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)
tree.score(X_test,y_test)
# Predict and evaluate
 
```

```python
#decision tree for white wine
tree = DecisionTreeClassifier(random_state=42)
tree.fit(X2_train, y2_train)
tree.score(X2_test,y2_test)
```

```python
#svc for red wine
from sklearn.svm import SVC
# Train the model
svm = SVC(C=6000)
svm.fit(X_train, y_train)
svm.score(X_test,y_test)
```

```python
#svc for white wine
svm = SVC(C=6000)
svm.fit(X2_train, y2_train)
svm.score(X2_test,y2_test)
```

```python
#random forest for red whine
from sklearn.ensemble import RandomForestClassifier
# Train the model
forest = RandomForestClassifier(n_estimators=100, random_state=42)
forest.fit(X_train, y_train)
print(forest.score(X_test,y_test))
y_pred = forest.predict(X_test)
```

```python
#random forest for white wine
forest = RandomForestClassifier(n_estimators=100, random_state=42)
forest.fit(X2_train, y2_train)
print(forest.score(X2_test,y2_test))
y2_pred = forest.predict(X2_test)
```

# Metrics

```python
#for red wine
# used metrics only for random forest as it got better score
#confusion matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
sns.heatmap(conf_matrix,annot=True)
plt.xlabel("predicted")
plt.ylabel("Truth")
plt.show()
```

```python
#classification report on random forest
print(classification_report(y_test, y_pred,zero_division=1))
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#now for white wine
conf2_matrix = confusion_matrix(y2_test, y2_pred)
print("Confusion Matrix:")
print(conf2_matrix)
```
